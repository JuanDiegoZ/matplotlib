```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

Object View

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y)
fig.show()
```

__Two Objects.__

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.9,0.9])
axes_two = fig.add_axes([0.2,0.5,0.3,0.4])
axes.plot(x,y,'bo')
axes_two.plot(y,x,'r')
axes_two.set_facecolor('blue')
fig.legend(["Hola Mundo"])
fig.show()
```
