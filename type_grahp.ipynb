__<h3>histogram<h3>__

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randint(1,50,100)
```

```python
plt.hist(data,bins = 20,histtype = 'step')
#numbers of bars. bins = x
plt.show
#Change Style. histtype = 'step'
```

```python
data = np.append(data, 200)
plt.boxplot(data,vert = False,patch_artist=True)
#Quit Outliers. Showfliers
plt.show
```

```python
data = np.append(data, 200)
plt.boxplot(data,vert = False,patch_artist=True,showfliers=False)
#Quit Outliers. Showfliers
plt.show
```

__<h2>Scatter<h2>__

```python
#load Data
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s=area)
plt.show()
```

```python
#Color and Marker.
plt.scatter(x,y, s=area, c= colors, marker = 'v') # cambiar el marker
plt.show()
```

```python
#Transparency
plt.scatter(x,y, s=area, c= colors, marker = 'p', alpha= 0.5) # transparencia
plt.show()
```
