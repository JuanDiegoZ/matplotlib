__Legendg, titles, more...__

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,5,11)
y = np.sin(x)** 2
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(8,4))
ax1.plot(x,y, 'b',label='$sin(x)$')
#Label in x axe
ax1.set_xlabel('X')
#label in y axe
ax1.set_ylabel('Y')
#Title of figure.
ax1.set_title('Relation X - Y')
#insert legend in a figure.
ax1.legend()
#Secon figure.
ax2.plot(y,x, 'r')
ax2.set_title('Relation Y - X')
plt.show()
```

__With pyplot__

```python
plt.plot(x,y, label='sin(x)')
plt.title('Un titulo comun')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc ='lower right')
plt.show()
```
