```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,6,11)
y = np.tan(x)
```

```python
fig,axes =  plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'r')
axes[1].plot(y,x,'b')
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3)
ax1.plot(y**2,x,'g')
ax2.plot(y,x,'g')
ax3.plot(y,x**3,'g')

```

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,y)
axes[0,1].plot(np.cos(y),x, 'r')
axes[0,2].plot(np.sin(y),x, 'r')
axes[0,3].plot(np.tan(y),x, 'r')
axes[1,0].plot(y,np.cos(x), 'r')
axes[1,1].plot(y,np.sin(x), 'r')
axes[1,2].plot(y,np.tan(x), 'r')
axes[1,3].plot(y,np.sqrt(x), 'r')
#mejorar la visualizacin de las figuras
fig.tight_layout()
```
