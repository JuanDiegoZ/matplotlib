```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
citys = ['Bogota','Mexico city','Berlin','Tokyo']
population = [2000,4000,900 ,3000]
```

```python
plt.bar(citys,population,width = 0.5,color = ['brown','green','blue','purple'])
plt.style.use('seaborn-v0_8')
plt.title('Population Cities')
#confiugration axe x.
plt.xlabel('Cities')
plt.xticks(np.arange(4),rotation= 45)
#configuration axe y
plt.ylabel('Population (Millions)')
plt.show(9)
```
