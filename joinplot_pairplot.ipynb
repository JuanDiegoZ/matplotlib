```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

 <h2>Joinplot<h2>

```python
exercise = sns.load_dataset('exercise')
exercise
```

```python
sns.jointplot(data=exercise, x ='pulse', y ='time');
```

```python

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue = 'sex');
```

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue='sex', kind= 'hist', marginal_ticks=True, #muestra una pequeÃ±a tabla para la grÃ¡fica exterior
marginal_kws=dict(bins= 25, fill = True, multiple= 'dodge'));
```

```python
sns.pairplot(data= tips, corner=True);
```
